<h1> EXTRACTING NECESSARY DATA FROM DATASET </h1<

```python
import pandas as pd
```

```python
df_players = pd.read_csv("FIFA_2022_Players.csv",low_memory = False)  #Reads the whole file at once
```

```python
df_players = df_players[['short_name', 'age', 'nationality_name', 'overall', 'potential',
         'club_name', 'value_eur', 'wage_eur', 'player_positions']]

# Selecting the most prefered positions of the players
df_players['player_positions'] = df_players['player_positions'].str.split(',' , expand = True)[0]
```

```python
players_missing_worldcup = ['K. Benzema', 'S. Mané', 'S. Agüero', 'Sergio Ramos', 'P. Pogba',
                            'M. Reus', 'Diogo Jota', 'A. Harit', 'N. Kanté', 'G. Lo Celso', 'Piqué',
                            'G. Wijnaldum','F. Wirtz', 'M. Maignan','Oyarzabal','T. Werner','R. James',
                            'Pedro Neto','B. Kamara','S. Kennedy','J. Corona','P. Kimpembe','Y. Nakayama',
                            'C. Nkunku','Gayà','M. Crépeau','J. Correa']
```

```python
delete_index = df_players[df_players['short_name'].isin(players_missing_worldcup)].index
df_players.drop(delete_index, axis = 0, inplace = True)
```

```python
worldcup_teams = ['Qatar', 'Brazil', 'Belgium', 'France', 'Argentina', 'England', 'Spain', 'Portugal',
    'Mexico', 'Netherlands', 'Denmark', 'Germany', 'Uruguay', 'Switzerland', 'United States', 'Croatia',
    'Senegal', 'Iran', 'Japan', 'Morocco', 'Serbia', 'Poland', 'South Korea', 'Tunisia',
    'Cameroon', 'Canada', 'Ecuador', 'Saudi Arabia', 'Ghana', 'Wales', 'Costa Rica', 'Australia']

df_players = df_players[df_players['nationality_name'].isin(worldcup_teams)]
```

```python
df_players.sort_values(by=['overall', 'potential', 'value_eur'], ascending=False, inplace=True)
df_players
```

<h1> BEST SQUAD FOR A TEAM </h1>

```python
def best_squad(team):
    df_best_squad = df_players.copy()
    df_best_squad = df_best_squad.drop(['value_eur','wage_eur'],axis = 1)
    df_best_squad = df_best_squad.groupby(['nationality_name','player_positions']).head(2)
    df_best_squad = df_best_squad[df_best_squad['nationality_name'] == team].sort_values(['player_positions','overall','potential'], ascending = False)
    return df_best_squad
```

```python
best_squad('Argentina')
```

<h1> TOP 5 PLAYERS OF THE TEAM </h1>

```python
def top5(team):
    df_top5 = df_players.copy()
    df_top5 = df_top5.drop(['value_eur','wage_eur'], axis = 1)
    df_top5 = df_top5[df_top5['nationality_name'] == team]
    df_top5 = df_top5.sort_values(['overall','potential'] , ascending = False).head()
    return df_top5
```

```python
top5('Spain')
```

<h1> OVERALL OF TEAMS </h1>

```python
avg_overall = [best_squad(team)['overall'].mean() for team in worldcup_teams]
```

```python
df_overall_teams = pd.DataFrame({'Team': worldcup_teams , 'Average Overall': avg_overall})
```

```python
df_overall_teams = df_overall_teams.dropna()
```

```python
df_overall_teams = df_overall_teams.sort_values('Average Overall', ascending = False)
```

```python
df_overall_teams
```

<h1> BEST TEAM FORMATION  </h1>

```python
dict_formation = {
    '4-3-3': ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'CM', 'CAM', 'RW', 'ST', 'LW'],
    '4-4-2': ['GK', 'RB', 'CB', 'CB', 'LB', 'RM', 'CM', 'CM', 'LM', 'ST', 'ST'],
    '4-2-3-1': ['GK', 'RB', 'CB', 'CB', 'LB', 'CDM', 'CDM', 'CAM', 'CAM', 'CAM', 'ST'],
}
```

```python
def best_lineup(nationality, lineup):
    lineup_count = [lineup.count(i) for i in lineup]

    df_lineup = pd.DataFrame({'position': lineup, 'count': lineup_count})
    positions_non_repeated = df_lineup[df_lineup['count'] <= 1]['position'].values
    positions_repeated = df_lineup[df_lineup['count'] > 1]['position'].values

    df_squad = best_squad(nationality)

    df_lineup = pd.concat([
        df_squad[df_squad['player_positions'].isin(positions_non_repeated)].drop_duplicates('player_positions', keep='first'),
        df_squad[df_squad['player_positions'].isin(positions_repeated)]]
    )
    return df_lineup[['short_name', 'overall', 'club_name', 'player_positions']]
```

```python
for index, row in df_overall_teams[:9].iterrows():
    max_average = None
    for key, values in dict_formation.items():
        average = best_lineup(row['Team'], values)['overall'].mean()
        if max_average is None or average>max_average:
            max_average = average
            formation = key
    print(row['Team'], formation, max_average)
```

<h2> CREATING DATAFRAMES FOR BEST SQUADS </h2>

```python
df_Spain = best_lineup('Spain', dict_formation['4-2-3-1'])
df_Portugal = best_lineup('Portugal', dict_formation['4-2-3-1'])
df_Brazil = best_lineup('Brazil', dict_formation['4-3-3'])
df_England = best_lineup('England', dict_formation['4-4-2'])
df_France = best_lineup('France', dict_formation['4-2-3-1'])
df_Argentina = best_lineup('Argentina', dict_formation['4-3-3'])
df_Germany = best_lineup('Germany', dict_formation['4-2-3-1'])
df_Belgium = best_lineup('Belgium', dict_formation['4-3-3'])
df_Netherlands = best_lineup('Netherlands', dict_formation['4-3-3'])
```

<h2> EXPORTING THE DATAFRAMES TO CSV FILES </h2>

```python
df_Spain.to_csv('Spain_Squad.csv', index = False)
df_Portugal.to_csv('Portugal_Squad.csv', index = False)
df_Brazil.to_csv('Brazil_Squad.csv', index = False)
df_England.to_csv('England_Squad.csv', index = False)
df_France.to_csv('France_Squad.csv', index = False)
df_Argentina.to_csv('Argentina_Squad.csv', index = False)
df_Germany.to_csv('Germany_Squad.csv', index = False)
df_Belgium.to_csv('Belgium_Squad.csv', index = False)
df_Netherlands.to_csv('Netherlands_Squad.csv', index = False)
```
